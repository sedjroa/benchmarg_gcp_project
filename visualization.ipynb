```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
def visualizer(dataset, title, xlabel, ylabel):
    data = dataset.groupby("PARAM").agg({"AVG_TIME":"mean"})
    sb.barplot(x="PARAM", y="AVG_TIME", data=data)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

### Données de concurrence utilisateurs, et de montée de charge

```python
conc = pd.read_csv("out/conc.csv", sep=",")
conc.head(2)
```

```python
post = pd.read_csv("out/post.csv", sep=",")
post.head(2)
```

```python
fanout = pd.read_csv("out/fanout.csv", sep=",")
fanout.head(2)
```

### Visualisation des données

```python
visualizer(conc, xlabel="Nombre d'utilisateurs concurrents",ylabel="Temps moyens d'exécution de requête",
                    title="Temps moyen par requête selon concurrence")
plt.savefig(f"res/conc.png", dpi=300, bbox_inches='tight')
```

```python
visualizer(post, xlabel="Nombre de postes par utilisateurs",ylabel="Temps moyens d'exécution de requête",
                    title="Temps moyen par requête selon concurrence")
plt.savefig(f"res/post.png", dpi=300, bbox_inches='tight')
```

```python
visualizer(fanout, xlabel="Nombre d'abonnement par utilisateurs",ylabel="Temps moyens d'exécution de requête",
                    title="Temps moyen par requête selon le nombre d'abonnement")
plt.savefig(f"res/fanout.png", dpi=300, bbox_inches='tight')
```
